```python
#importing the necessary libraries
import numpy 
import cv2
import os 
import pickle
from sklearn.svm import LinearSVC
from sklearn.calibration import CalibratedClassifierCV
```

```python
#Loading in the cnn
net = cv2.dnn.readNetFromCaffe("./bvlc_alexnet.prototxt", "./bvlc_alexnet.caffemodel")
```

```python
# Inspecting the cnn's architecture
layer_names = net.getLayerNames()
print(layer_names)
```

```python
def extract_feature(filename, foldername):
    print(filename)
    image = os.path.join('JPEGImages', foldername, filename)
    img = cv2.imread(image)
    #resizing and normailizing the image data to reach the desired state for the cnn
    # (1, 3, 224, 224)
    resized = cv2.resize(img,(224, 224), interpolation = cv2.INTER_CUBIC)
    blob = cv2.dnn.blobFromImage(resized, 1, (224, 224), (104, 117, 123))
    net.setInput(blob)

    # To get the feature representations of the training and test images, 
    # use activations on the fully connected layer 'fc7'
    features = net.forward("fc7")
    return features
```

```python
#creating the necessary directory strings
cwd = os.getcwd()
imfold = os.path.join(cwd,"JPEGImages")
#ignoring the .ds_store files in the JPEGImages folder
dirlist = os.listdir(imfold)[1:]
```

```python
#in this cell I create the extracted feature vectors for each class separately to help generate the y data later on

for folder in dirlist:
    extracted = []
    #creating a list of the image names present in the current folder
    currdir = os.path.join(imfold, folder)
    imlist = os.listdir(currdir)
    #skipping the .ds_store file and the extracted if re-extracting features
    for file in imlist:
        if file == "extracted.txt" or file == ".DS_Store":
            continue
        extracted.append(extract_feature(file, currdir))
    
    extracted = numpy.array(extracted)
    sh = extracted.shape
    extracted = extracted.reshape((sh[0], sh[2]))
    #checking the size of each folder's extracted vectors for imbalances in the data
    print(extracted.shape)
    
    numpy.savetxt(X=extracted, fname=os.path.join(currdir,"extracted.txt"))
```

```python
#loading in the necessary data
attvect = numpy.loadtxt('predicate-matrix-binary.txt')
classes = list(numpy.loadtxt('classes.txt', delimiter='\t', usecols=[1], dtype=numpy.str))
trainlist = list(numpy.loadtxt('trainclasses.txt', delimiter='\n', dtype=numpy.str))
testlist = list(numpy.loadtxt('testclasses.txt', delimiter='\n', dtype=numpy.str))
```

```python
'''
In this cell I create two vectors to store the predicate vectors 
for the training and testing classes, respectively.
'''
print(trainlist)

print("-"*120)

train_encs = numpy.array([attvect[classes.index(c), :] for c in trainlist])
print(train_encs)

print("-"*120)

print(testlist)

print("-"*120)

test_encs = numpy.array([attvect[classes.index(c),:] for c in testlist])
print(test_encs)
```

```python
#in this function I create the input matrix in the appropriate format and the entry list which will be used in ydatagen
def xdatagen(categlist):
    X = list()
    entrylist = list()
    for folder in categlist:
        currdir = os.path.join(imfold,folder)
        #load extracted features
        ext = numpy.loadtxt(os.path.join(currdir, "extracted.txt"))
        #getting number of entries and appending to list to use in y data matrix population
        entries = ext.shape[0]
        entrylist.append(entries)

        for ind in range(entries):
            X.append(ext[ind,:])
        #print("entered "+str(entries)+" from folder "+folder) // this was used to monitor the creation of the data
    return X, entrylist
```

```python
#creating the training input data in the appropriate format
Xtrain, trainentries = xdatagen(trainlist)
print(len(Xtrain))
print(trainentries)
```

```python
#creating the testing input data in the appropriate format
Xtest, testentries = xdatagen(testlist)
print(len(Xtest))
print(testentries)
```

```python
#here I use the entrylist with the appropriate class list to format the y data appropriately
def ydatagen(categlist, entrylist):
    ymat = list()
    for i, entries in enumerate(entrylist):
        for cnt in range(entries):
            ymat.append(attvect[classes.index(categlist[i]),:])
    return numpy.array(ymat)
```

```python
#in this cell, the y data for training and testing are created 
ytrain = ydatagen(trainlist, trainentries)
print(len(ytrain))
print(ytrain)
ytest = ydatagen(testlist, testentries)
print(len(ytest))
print(ytest)
```

```python
#here I'll create and train the attribute prediction models
def trainSVC(att):
    svm = LinearSVC(C=1e-5)
    clf = CalibratedClassifierCV(svm)
    clf.fit(Xtrain, ytrain[:, att])
    print("accuracy measured to be: {}".format(clf.score(Xtest, ytest[:, att])))
    return clf 

def train_attribute_models(func):
    clflist = []
    for i in range(attvect.shape[1]):
        print("Starting Support Vector Classifier training on classifier #{}".format(i+1))
        clflist.append(func(i))
    return clflist
clflist = train_attribute_models(trainSVC)
```

```python
#saving the classifiers to a pickle file to skip training for easier work throughout the days of development
pickle_out = open("RFCs.pickle","wb")
pickle.dump(clflist, pickle_out)
pickle_out.close()
```

```python
#loading said models
pickle_in = open("RFCs.pickle", "rb")
clflist = pickle.load(pickle_in)
```

```python
'''
In this function I diverged from the instructions slightly and return the probabilities for 0 and for 1
since the class probability calculation example used both the probability of an attribute existing i.e. P(att=1|x) 
and not existing P(att=0|x) so both probabilities will be useful in the next part.
Of course if only the probabilities of the attribute existing are needed we can use 
probslist = []
for ext in X[test]:
    probslist.append(compute_attribute_probs(ext)[:,1])
return probslist
    
'''
#getting the probabilities for every attribute of a vector
def compute_attribute_probs(ext):
    return numpy.array([clflist[i].predict_proba([ext])[0] for i in range(attvect.shape[1])])

proba_vect = compute_attribute_probs(Xtest[0])
print(proba_vect.shape)
print(proba_vect)

```

```python
def compute_class_probs(proba_vect):
    #list with probability for each class
    problist = list()
    for i in range(test_encs.shape[0]):
        enc = test_encs[i,:]
        prob = 1
        for ind in range(enc.shape[0]):
            prob*=proba_vect[ind, int(enc[ind])]
        problist.append(prob)
    return problist        
```

```python
#testing the function's output
prediction = compute_class_probs(proba_vect)
print(prediction)
prediction = numpy.argmax(prediction)
print(prediction)
```

```python
#computing the average accuracy of prediction
def compute_accuracy():
    count = 0
    for i in range(len(ytest)):
        probavect = compute_attribute_probs(Xtest[i])
        probs = compute_class_probs(probavect)
        prediction = numpy.argmax(probs)
        if numpy.array_equal(ytest[i], test_encs[prediction]):
            count+=1
    return float(count)/ len(ytest)
print("Final accuracy is: {}%".format(round(compute_accuracy()*100, 3)))
```
